```python
import pandas as pd
import mysql.connector
import os

# Define CSV files and corresponding table names
csv_files = [
    ('Products.csv', 'products'),
    ('Customers.csv', 'customers'),
    ('Location.csv', 'location'),
    ('Orders.csv', 'orders')
]

# Folder containing the CSVs
folder_path = r"C:\Users\garga\Downloads\EU-Dataset"

# Connect to MySQL
conn = mysql.connector.connect(
    host='localhost',
    user='root',
    password='admin',  # Replace if needed
    database='Sales_data',
    auth_plugin='mysql_native_password'
)
cursor = conn.cursor()

# Function to map pandas data types to MySQL
def get_sql_type(dtype):
    if pd.api.types.is_integer_dtype(dtype):
        return 'INT'
    elif pd.api.types.is_float_dtype(dtype):
        return 'FLOAT'
    elif pd.api.types.is_bool_dtype(dtype):
        return 'BOOLEAN'
    elif pd.api.types.is_datetime64_any_dtype(dtype):
        return 'DATETIME'
    else:
        return 'TEXT'

# Loop through each CSV file
for csv_file, table_name in csv_files:
    file_path = os.path.join(folder_path, csv_file)
    print(f"Processing {csv_file}...")

    # Read CSV with appropriate encoding and delimiter
    try:
        df = pd.read_csv(file_path, encoding='latin1', sep=';')
    except Exception as e:
        print(f"Error reading {csv_file}: {e}")
        continue

    # Replace NaN with None
    df = df.where(pd.notnull(df), None)

    # Clean column names
    df.columns = [col.strip().replace(' ', '_').replace('-', '_').replace('.', '_') for col in df.columns]

    # Auto-generate CREATE TABLE query
    columns = ', '.join([f'`{col}` {get_sql_type(df[col].dtype)}' for col in df.columns])
    create_table_query = f"CREATE TABLE IF NOT EXISTS `{table_name}` ({columns})"
    try:
        cursor.execute(create_table_query)
    except Exception as e:
        print(f"Error creating table {table_name}: {e}")
        continue

    # Insert rows
    for _, row in df.iterrows():
        values = tuple(None if pd.isna(x) else x for x in row)
        placeholders = ', '.join(['%s'] * len(row))
        cols = ', '.join([f'`{col}`' for col in df.columns])
        sql = f"INSERT INTO `{table_name}` ({cols}) VALUES ({placeholders})"
        try:
            cursor.execute(sql, values)
        except Exception as e:
            print(f"Error inserting row into {table_name}: {e}")
            continue

    # Commit after each file
    conn.commit()
    print(f"✅ Successfully loaded {csv_file} into `{table_name}`\n")

# Close MySQL connection
cursor.close()
conn.close()
print("✅ All files processed and database connection closed.")
```
